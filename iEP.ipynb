```python
bins=[0,18,36,72]
labels=["child","adult","old"]
df["agegroup"]=pd.cut(df["Age"],bins=bins,labels=labels)
```

```python
df.groupby("agegroup")("Survived").mean()*100
```

 4 data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x="Sex",y="Survived",data=df,color="green",edgecolor="black")
plt.title("no.of males and female survived")
plt.xlabel("gender")
plt.ylabel("survival rate")
plt.show()
```

```python
sns.barplot(x="Pclass",y="Survived",data=df,color="yellow",edgecolor="black")
plt.title("No. of Passengers Survived")
plt.xlabel("Passnger Class")
plt.ylabel("Survival rate")
plt.show()
```

```python
sns.barplot(x="age_group",y="Survived",data=df,color="orange",edgecolor="black")
plt.title("agegroup vs Survival rate")
plt.xlabel("agegroup")
plt.ylabel("Survival rate")
plt.show()
```
